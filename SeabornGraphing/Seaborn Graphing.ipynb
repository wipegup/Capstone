showPlot is a general graphing function for creating heatmaps of sequenceRank v. dateRank data.

It has several options:  
- source *unactivated*  


- kind : 'agg'(default) or 'unagg' 'agg' creates visualizations which aggregates all members of each (sequenceRank, dateRank) pair. 'unagg' creates heatmaps for all species per Genus from 1 to 10  
  
  
- num: Only relevant for 'agg' plots, either 1 or 'All', if 'All' Shows all heatmaps created from options, if 1, only creates heatmap of "chosen" options  

- spgDiscard: a list ([1] is default) connotes which species per genus to disregard (only relevant for agg plots)  


- func: 'Prop'(default), 'Count', 'AbsZ', 'pVal' chooses which function to use when calculating heat- map values: 'prop': observed species / expected species. 'count': observed species. 'z': absolute z-value. 'pVal', pVal calculated from normal distribution  


- SDateVGDate: None (default) True, or False, creates split comparing Species date of description to Genus date of erection, comparison dependant upon 'SGvsOperator'  
- SGvsOperator : '>' (default), '>=', '<', '<=', '==' defines the comparison for the Species description date, and Genus date of erection.  

For example, "SDateVGDate = True, SGvsOperator = '<'" will return a visualization where the species date of description is less than (exclusive) of genus erection date. Will return the same as "SDateVGDate = False, SGvsOperator = '>='"  


- Type : None (default), True, False. Specifies splitting between type-species and non-type species.  


- OrigGenus : None(default), True, False. Specifies a split between taxa in original genus, and not in original genus.  


- Epoch : None (default), True, list. If True, splits, into epochs of 1750-1815, 1815-1915, 1915-present. Which is equivalent to Epoch = [1758, 1815, 1915, 2020]. If only one graphic shown, only data from first epoch is used.  


- EpochType : 's' (default) or 'g'. Determines whether taxa-epochs are defined by species description date or genera erection date  


Note: When an option is 'None', no split will be created, but if something is 'False' the split will be created
